# Plote um gráfico de barras (bar) que exiba a quantidade de funcionários em cada setor (coluna Department);
Resposta: A quantidade de funcionários em (Research & Development) é superior a quantidade de funcionários em (Human Resources) com grande significancia

```python
df["Department"].value_counts().plot(kind="bar", color=["blue", "cyan", "green"])
plt.xticks(rotation=45)
plt.show()
```

#  Plote um gráfico de linha (line) que exiba a quantidade de funcionários para cada nível de escolaridade (coluna Education);
Resposta: O nivel de escolariedade medio entre os funcionários é entre 3 e 3.5


```python
education_counts = df['Education'].value_counts()
```

```python
plt.plot(education_counts.values, education_counts.index, marker='o', linestyle='-', color='blue')
plt.xlabel("Quantidade de Pessoas")
plt.ylabel("Nível de Educação")
plt.title("Distribuição dos Níveis de Educação")
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

#  Plote um gráfico de torta/pizza (pie) que exiba a quantidade de funcionários que tem ou não algum atrito na empresa (coluna Attrition).
Resposta: 16.1% dos funcionarios que responderam a pesquisa, possuem algum tipo de atrito no ambiente de trabalho

```python
atrito = df['Attrition'].value_counts()
cores = sns.color_palette("pastel", len(atrito))

plt.pie(atrito, autopct='%1.1f%%', colors=cores, startangle=90, explode=(0.05, 0))
plt.title("Distribuição do Atrito no Trabalho", pad=20)

novos_rotulos = {'No': 'Não tem atrito', 'Yes': 'Sim tem atrito'}
plt.legend(labels=[novos_rotulos[label] for label in atrito.index], loc="best")
plt.axis("equal")

plt.show()
```
